```python
#in a sense re-keying the dictionary for season highs
#mismatch between the columns for the type of statistic and the value of the stat
for x in range(0,14):
    s = df.stats[x]['gamehigh_stats']['Season Highs'][1:]
    for y in range(0,14):
        try:
            t=s[y]['value']
            s[y]['statistic'] = t
            s[y]['value'] = s[y]['opponent']
            del s[y]['opponent']
        except IndexError:
            continue
```
